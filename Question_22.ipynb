#### Extract the strings from the strings
#### how many time reapeted abc in x string

```python
x="abcdcdc" 
y="cdc"
count=0
for i in range(len(x)):
	for j in range(len(y)):
		if x[i+j]==y[j] and j==len(y)-1:
			count+=1
		if x[i+j]!=y[j]:
			break
	if i==len(x)-len(y):
		break
print(count)
```
